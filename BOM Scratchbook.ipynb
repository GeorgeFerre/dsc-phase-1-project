```python
bom_dropped = bom_df.dropna()
bom_dropped.info()
```

```python
bom_dropped['foreign_gross'] = bom_dropped['foreign_gross'].apply(lambda x: float(x.split()[0].replace(',', '')))
bom_dropped['foreign_gross'] = bom_dropped['foreign_gross'].astype(float)
bom_dropped.head()
```

```python
bom_dropped['total_gross_millions'] = (bom_dropped['domestic_gross'] + bom_dropped['foreign_gross'])/1000000
bom_dropped.head()
```

```python
bom_dropped['total_gross_millions'].describe()
```

```python
bom_dropped.max()
```

```python
bom_dropped = bom_dropped.sort_values(by='total_gross_millions', ascending=False)
bom_dropped.head(20)
```

```python
bom_dom_greater = bom_dropped[bom_dropped['domestic_gross'] > bom_dropped['foreign_gross']]
bom_dom_greater.head(20)
```
